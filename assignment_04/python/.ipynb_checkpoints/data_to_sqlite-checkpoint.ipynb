### Step 1: Import Dataset From File

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_json.html

```python
import pandas as pd
import numpy as np
import sqlite3
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
# Column names from wine.names
columns = [
    'cultivar', 'alcohol', 'malic_acid', 'ash', 'alcalinity_of_ash',
    'magnesium', 'total_phenols', 'flavanoids', 'nonflavanoid_phenols',
    'proanthocyanins', 'color_intensity', 'hue', 'od280_od315', 'proline'
]

# Read wine.data (no header, comma-separated)
wine = pd.read_csv('wine.data', header=None, names=columns)
wine.head()
```

```python
# Some statistics
wine.describe()
```

```python
# look at first few rows
wine.head()
```

```python
# Check class distribution
print(wine['cultivar'].value_counts())
# Map cultivar number to readable name
wine['cultivar_name'] = wine['cultivar'].map({1: 'Cultivar 1', 2: 'Cultivar 2', 3: 'Cultivar 3'})
```

### Step 2: Dimensionality Reduction

For older apporaches see https://scikit-learn.org/stable/index.html
For UMAP see https://github.com/lmcinnes/umap

```python
# Selecting the 13 chemical feature columns (exclude cultivar label)
feature_cols = [
    'alcohol', 'malic_acid', 'ash', 'alcalinity_of_ash', 'magnesium',
    'total_phenols', 'flavanoids', 'nonflavanoid_phenols',
    'proanthocyanins', 'color_intensity', 'hue', 'od280_od315', 'proline'
]

# Standardize features (mean=0, std=1) — important before PCA
scaler = StandardScaler()
X_scaled = scaler.fit_transform(wine[feature_cols])
# Apply PCA to reduce to 2 components
pca = PCA(n_components=2)
pca_coords = pca.fit_transform(X_scaled)

wine['pca_x'] = pca_coords[:, 0]
wine['pca_y'] = pca_coords[:, 1]

print(f"Explained variance ratio: PC1={pca.explained_variance_ratio_[0]:.3f}, PC2={pca.explained_variance_ratio_[1]:.3f}")
print(f"Total variance explained: {sum(pca.explained_variance_ratio_):.3f}")
```

### Step 3: Clustering

See https://scikit-learn.org/stable/index.html

```python
# KMeans with 3 clusters (matching the 3 cultivars)
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
wine['cluster'] = kmeans.fit_predict(X_scaled)

print("Cluster distribution:")
print(wine['cluster'].value_counts())
```

### Step 4: Write Results to Database

Write the orginal data (the part you kept) along with the dimensionality recudction coordinates and the cluster IDs to the database


Open the database

https://docs.python.org/3.9/library/sqlite3.html (pay attention to your python version)


```python
connection = sqlite3.connect("./data.db")
```

Write to database 

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.to_sql.html#pandas.DataFrame.to_sql

```python
# Write to database — table name 'wine'
# Note: server.py queries the 'iris' table by default, so we update that too
wine.to_sql("wine", con=connection, if_exists='replace', index_label='id')
# Also write as 'iris' so existing server.py works without changes
wine.to_sql("iris", con=connection, if_exists='replace', index_label='id')
print(f"Saved {len(wine)} rows to data.db")
print(f"Columns: {wine.columns.tolist()}")
```

Close database connection

```python
connection.close()
print("Done!")
```
