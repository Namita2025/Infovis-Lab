# Column Descriptions

## Identifier Columns
**ID** (object): Unique identifier for each train ride entry. Contains 105,728 unique values 
out of 106,555 rows, meaning some trains appear at multiple stations. 
Format: numeric-string combined with date information.

**eva_nr** (int64): Official DB station number (EVA number). 
76 unique values corresponding to 76 stations in the VVS network. 
Range: 8000016 to 8006743.

## Train Information
**line** (object): The train line identifier. 29 unique lines exist including 
S-Bahn lines (1=S1, 2=S2 etc.), regional lines (RB46, RB47) and other 
services (17a, 17c, 60, 90). No missing values.

**path** (object): Full route path of the train as pipe-separated station names. 
8,783 missing values (trains at start/end of route). 
583 unique route combinations exist.

**category** (int64): Station size/importance category ranging from 1 (largest) 
to 5 (smallest). Mean of 3.59 indicates most stations are medium-small sized. 
No missing values.

## Location Columns
**station** (object): Name of the station where the train arrives. 
76 unique stations all within the VVS Stuttgart S-Bahn network. No missing values.

**state** (object): German federal state. Only one unique value: 
Baden-Wuerttemberg, since all stations are within the VVS network.

**city** (object): City name where the station is located. 
45 unique cities — fewer than stations because some cities have multiple stations.

**zip** (int64): Postal code of the station location. 
59 unique values, range 70173 to 76314, covering the Stuttgart metro area.

**lat** (int64): Latitude coordinate of the station stored without decimal point. 
Must be divided by 1,000,000 for correct geographic coordinates 
(e.g. 48720393 → 48.720393). 76 unique values, one per station.

**long** (int64): Longitude coordinate stored without decimal point. 
Must be divided by 1,000,000 for correct geographic coordinates 
(e.g. 9382384 → 9.382384). 76 unique values, one per station.

**previous_nr** (int64): EVA number of the previous station in the train's route. 
75 unique values — used to reconstruct network connections between stations.

**previous_station** (object): Name of the previous station in the train's route. 
75 unique values. Special values include "Start" (train begins here) and 
"Station outside VVS" (train comes from outside the network).

## Time Columns
**arrival_plan** (object): Planned arrival time at the station in format DD.MM.YYYY HH:MM. 
8,783 missing values — these are departure-only stops (first station on route).

**departure_plan** (object): Planned departure time at the station. 
No missing values since every entry has a departure time.

**arrival_change** (object): Actual/updated arrival time after delays. 
17,840 missing values — more than arrival_plan because some trains 
had no real-time tracking data available.

**departure_change** (object): Actual/updated departure time after delays. 
11,497 missing values — real-time departure updates not always available.

## Delay Columns
**arrival_delay_m** (int64): Arrival delay in minutes. Range 0-74 minutes, 
mean of 1.28 minutes indicating most trains arrive nearly on time. 
55 unique delay values. No missing values.

**departure_delay_m** (int64): Departure delay in minutes. Range 0-75 minutes, 
mean of 1.45 minutes — slightly higher than arrival delay suggesting 
delays sometimes increase at stations. 56 unique values. No missing values.

**arrival_delay_check** (object): Categorical flag — either "on_time" or "delay". 
Derived from arrival_delay_m. No missing values.

**departure_delay_check** (object): Categorical flag — either "on_time" or "delay". 
Derived from departure_delay_m. No missing values.

## Status Column
**info** (object): Reason for disruption if applicable. Only 3 unique values: 
"Information", "Stoerung" (disruption), "Bauarbeiten" (construction work). 
26,699 missing values meaning most trains run without any reported issues.

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1. Full distribution (log scale to see the tail)
df['arrival_delay_m'].plot(kind='hist', bins=50, ax=axes[0][0], 
                           title='Arrival Delay (log scale)', color='steelblue')
axes[0][0].set_yscale('log')
axes[0][0].set_xlabel('Minutes')

df['departure_delay_m'].plot(kind='hist', bins=50, ax=axes[0][1], 
                              title='Departure Delay (log scale)', color='coral')
axes[0][1].set_yscale('log')
axes[0][1].set_xlabel('Minutes')

# 2. Only delayed trains (delay > 0) - more interesting!
delayed = df[df['arrival_delay_m'] > 0]
delayed['arrival_delay_m'].plot(kind='hist', bins=40, ax=axes[1][0],
                                 title=f'Arrival Delay > 0 ({len(delayed)} trains)', 
                                 color='steelblue')
axes[1][0].set_xlabel('Minutes')

delayed2 = df[df['departure_delay_m'] > 0]
delayed2['departure_delay_m'].plot(kind='hist', bins=40, ax=axes[1][1],
                                    title=f'Departure Delay > 0 ({len(delayed2)} trains)', 
                                    color='coral')
axes[1][1].set_xlabel('Minutes')

plt.tight_layout()
plt.show()

# Key stats
print(f"Trains with arrival delay > 0: {len(delayed)} ({len(delayed)/len(df)*100:.1f}%)")
print(f"Trains with departure delay > 0: {len(delayed2)} ({len(delayed2)/len(df)*100:.1f}%)")
print(f"\nAverage arrival delay (all trains): {df['arrival_delay_m'].mean():.2f} min")
print(f"Average departure delay (all trains): {df['departure_delay_m'].mean():.2f} min")
print(f"\nAverage arrival delay (delayed only): {delayed['arrival_delay_m'].mean():.2f} min")
print(f"Average departure delay (delayed only): {delayed2['departure_delay_m'].mean():.2f} min")
```

```python
# Fix coordinates first
df['lat'] = df['lat'] / 1000000
df['long'] = df['long'] / 1000000

# Calculate delay_change per train
df['delay_change'] = df['departure_delay_m'] - df['arrival_delay_m']

# Overall delay_change per station
station_stats = df.groupby('station').agg(
    delay_change_avg = ('delay_change', 'mean'),
    arrival_delay_avg = ('arrival_delay_m', 'mean'),
    departure_delay_avg = ('departure_delay_m', 'mean'),
    sum_trains = ('ID', 'count')
).reset_index()

print("Station delay stats:")
print(station_stats.sort_values('delay_change_avg', ascending=False).head(10))
print("\nBiggest AMPLIFIERS (delay gets worse):")
print(station_stats.nlargest(5, 'delay_change_avg')[['station','delay_change_avg']])
print("\nBiggest ABSORBERS (delay gets better):")
print(station_stats.nsmallest(5, 'delay_change_avg')[['station','delay_change_avg']])
```

```python
# S-Bahn lines only
sbahn_lines = ['1','2','3','4','5','6','60','90']
df_sbahn = df[df['line'].isin(sbahn_lines)]

# delay_change per station per line
line_stats = df_sbahn.groupby(['station','line']).agg(
    delay_change_avg = ('delay_change', 'mean'),
    arrival_delay_avg = ('arrival_delay_m', 'mean'),
    departure_delay_avg = ('departure_delay_m', 'mean'),
    sum_trains = ('ID', 'count')
).reset_index()

print("Per line delay stats sample:")
print(line_stats.head(15))

print("\nHauptbahnhof per line:")
print(line_stats[line_stats['station'] == 'Hauptbahnhof'])
```
